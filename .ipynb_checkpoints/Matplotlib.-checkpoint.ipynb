```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as mt
```

## Basic Graphs.

```python
x = (1,2,3)
y = (1,2,3)


plt.plot(x,y, label="2x", color="pink",linestyle="--" ,linewidth = 1.5, marker = "*", markeredgecolor= "red")
plt.title("Our first Graph",fontdict={"fontsize":22,"fontname":"comic sans ms"})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show(x,y)
```

```python

plt.subplot(1,2,1)
plt.plot(x,y,color="orange", linestyle=":", linewidth = 5)
plt.title("Straight line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,z,color="green", linestyle="-.", linewidth = 3)
plt.title("Straight line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
